```python
x,y=image.shape[0:2]
plot.imshow([[sum(image[i,j])/3 for j in range(y)] for i in range(x)])
```

```python
p,q=25,100
c=2
def dot(x,y):
    if math.sqrt((x-p)**2 + (y-q)**2)<=c:
        return [255,0,0]
    else:
        return conv[x,y]
plot.imshow([[dot(i,j) for j in range(y)] for i in range(x)])
```

```python
X,Y=image.shape[0:2]
cap = cv.imread('cap.png')
small_cap = cv.resize(cap, (100,30))
```

```python
ni=image.copy()
xs = 70
ys = 0
ni[ys:ys+small_cap.shape[0], xs:xs+small_cap.shape[1]] = small_cap
plot.imshow(cv.cvtColor(ni,cv.COLOR_BGR2RGB))
```

```python
p,q=30,100
c=2
xs = 55
ys = 0
def imadd(y,x):
    if math.sqrt((y-p)**2 + (x-q)**2)<=c:
        return image[25,0]
    if ys<=y<ys+small_cap.shape[0] and xs<=x<xs+small_cap.shape[1] and all(small_cap[y-ys,x-xs]!=[247, 247, 247]):
        return small_cap[y-ys,x-xs]
    else:
        return image[y,x]
cap_panda=np.array([[imadd(i,j) for j in range(Y)] for i in range(X)])
cv.imwrite('Images/cap_panda.jpg',cap_panda)
cap_panda=cv.cvtColor(cap_panda,cv.COLOR_BGR2RGB)
plot.imshow(cap_panda)
```
